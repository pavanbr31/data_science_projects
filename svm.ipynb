```python
x = DF.drop(columns = 'Outcome',axis=1)
y = DF['Outcome']
print(x)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(x)
```

```python
sd = ss.transform(x)
#print(sd)
```

```python
from sklearn.model_selection import train_test_split
x_tr,x_te,y_tr,y_te=train_test_split(x,y,test_size=0.2,stratify=y,random_state=5)
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel='linear',random_state=0)
classifier.fit(x_tr,y_tr)
```

```python
y_pred = classifier.predict(x_tr)
#print(y_pred)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
acc=accuracy_score(y_pred,y_tr)
cm = confusion_matrix(y_pred,y_tr)
print(cm)
print(acc)
```

```python
x_pred1 = classifier.predict(x_te)
#print(x_pred1)
```

```python
acc1=accuracy_score(x_pred1,y_te)
print(acc1)
```

```python
from mlxtend.plotting import plot_confusion_matrix,plot_decision_regions
plot_confusion_matrix(cm)
plt.show()
```
